```python
import pandas as pd
import numpy as np
a = pd.read_excel('./아파트(매매)_실거래가_20240227162710.xlsx')
a = a.drop(index=range(0,11,1))
a.columns = a.loc[11,:]
a.drop([11], axis=0, inplace=True)
a.to_csv('./Seoul_Apartment.csv', index=False)
```

```python
class csv_control:
    def __init__(self, df):
        self.df = df
    def check(self):
        display(self.df.columns)
        display(self.df.info())
        display(self.df.isnull().sum())
    def data_setting(self, col_list):
        self.df = self.df.drop(col_list, axis=1)
        display(self.df.info())
    def columns_compare(self, compare_df):
        list1 = list(set(self.df.columns))
        list2 = list(set(compare_df.columns))
        list1.sort()
        list2.sort()
        display(list1)
        display(list2)
        display(list1==list2)

Seoul = csv_control(pd.read_csv('./Seoul_Apartment.csv'))
```
